```python
import numpy as np
import pandas as pd
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer, TfidfTransformer, TfidfVectorizer
from datasets import load_dataset
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import recall_score, f1_score, accuracy_score
```

```python
# Load data
df = load_dataset("stanfordnlp/imdb")
```

```python
train_data = df["train"]
test_data = df["test"]
train_df = pd.DataFrame(df["train"])
test_df = pd.DataFrame(df["test"])
test_df , _= train_test_split(test_df, test_size=0.8, random_state=42,)
```

```python

LR_pipeline = Pipeline([
        ('vect', CountVectorizer()),
         ('clf',  LogisticRegression()),
        ])
LR_pipeline.fit(train_df['text'], train_df['label'])
```

```python
y_pred = LR_pipeline.predict(test_df['text'])
```

```python
print('Baseline Model Performance using BOW feature extraction on Logistic regression')
print('Accuracy: ', accuracy_score(test_df['label'], y_pred))
```

```python
pipeline_MNB = Pipeline([
        ('vect', CountVectorizer()),
         ('clf',  MultinomialNB()),


    ])
pipeline_MNB.fit(train_df['text'], train_df['label'])
```

```python
y_pred = pipeline_MNB.predict(test_df['text'])
```

```python
print('Baseline Model Performance using BOW feature extraction on Multinomial bayes')
print('Accuracy: ', accuracy_score(test_df['label'], y_pred))
```
